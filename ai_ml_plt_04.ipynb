```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Movies vs Oscar Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years, oscar_revenue, color="red", marker="o", linestyle = "--")
plt.plot(non_oscar_years, non_oscar_revenue, color="green", marker="o", linestyle="--")
plt.title("Oscar vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.show()
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, color = "green", label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, color = "red", label="Non Oscar")

plt.title('Yearly revenue of Oscar and Non Oscar Movies')
plt.xlabel("Years")
plt.ylabel("Revenue in $M")
plt.legend()
plt.tight_layout()
plt.xticks(x, years)
plt.show()
```

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

colors = ["green" if x<135 else "red" for x in blood_pressure]
sizes = blood_pressure
plt.scatter(age, blood_pressure, color=colors, s=sizes, marker="o", alpha = 0.8 )

plt.scatter(age, blood_pressure, c = blood_pressure, cmap="OrRd")
plt.colorbar(label="BP")

plt.show()
```
