```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data 
```

```python
data = load_data("amazon_store_data.json")
```

```python
print(data)
```

```python
#function to clean the data
def clean_data(data):
    text_to_num = {"one" : 1, "two" : 2, "three" : 3, "four": 4, "five": 5}
    
    unique_users = set()
    cleaned_data = []
    
    for user in data:
        #clean rating - consistent typing
        raw_rating = str(user['rating']).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user['rating'] = raw_rating

        #handle missing values
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #handle duplicate values - deduplication
        if(user['name'].strip() in unique_users):
            continue
        unique_users.add(user['name'])
        cleaned_data.append(user)
        
    return cleaned_data

    
```

```python
data = clean_data(data)
print(data, len(data))
```

```python
# function to get the average rating of the product..
def get_insight(data):
    
    total_rating = 0
    len_rating = 0
    for user in data:
        total_rating += float(user['rating'])
        len_rating +=1
        
    print(f"Average rating of the Produt is {total_rating/len_rating}")
```

```python
get_insight(data)
```

```python
# To calculate the percetage of bad rating on product...
def percent_rating(data):

    good_rating = 0
    bad_rating = 0

    for user in data:
        if (float(user['rating']) > 3):
            good_rating += 1
        else:
            bad_rating += 1

    print(f"The percentage of bad rating is {int(bad_rating/(good_rating + bad_rating)*100)}%")
```

```python
percent_rating(data)
```

```python
# reccomment the produnt on the rating of customer 

def rec_product(data):

    same_product = []
    diff_product = []

    for user in data:
        if (float(user['rating']) >= 4):
            same_product.append(user['name'])
        else:
            diff_product.append(user['name'])

    print(f"Recomment same product of users : {same_product} \nRecomment the Different product of users : {diff_product}")
    
```

```python
rec_product(data)
```
